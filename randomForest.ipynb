```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
from sklearn.model_selection import KFold,cross_val_score
```

```python
def r2_scorer(estimator, X, y):
    predictions = estimator.predict(X)
    return r2_score(y, predictions)
```

```python
train = pd.read_csv('abalone.csv')
train['Sex'] =  train['Sex'].map(lambda x: 1 if x == 'M' else (-1 if x == 'F' else 0))
y = train.Rings  
X = train.drop(['Rings'], axis=1)
```

```python
scores = []
validator = KFold(y.size, n_splits=5, shuffle=True, random_state=1)
for n_est in range(1, 51):
    clf = RandomForestRegressor(random_state=1, n_estimators=n_est)
    clf.fit(X, y)
    scores.append((
        n_est,
        cross_val_score(clf, X, y = y, cv = validator, scoring=lambda est, X, y: r2_score(y, est.predict(X))).mean()
    ))
print(scores)
```
